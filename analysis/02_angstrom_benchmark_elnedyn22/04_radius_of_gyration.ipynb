````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os

dir_names = "baseline all_common 0.1nm 0.05nm 0.025nm 0.01nm no_rubber".split(" ")
label_names = ["Baseline", "No threshold", "Threshold: 0.1 nm", "Threshold: 0.05 nm", "Threshold: 0.025 nm" , "Threshold: 0.01 nm"]

df = pd.DataFrame()
for protein in ["5eqi", "4zw9"]:
    for threshold in dir_names:
        for i in range(3):
            path = f"gr/{protein}_{threshold}_rep{i+1}.xvg"
            if not os.path.isfile(path):
                continue

            x, y, rg_x, rg_y, rg_z = np.loadtxt(path,comments=["@", "#"],unpack=True)
            res = pd.DataFrame({
                "time" : x,
                "rg_all" : y,
                "rg_x"   : rg_x,
                "rg_y"   : rg_y,
                "rg_z"   : rg_z,
                "replicate" : f"rep {i+1}",
                "protein"   : protein,
                "threshold" : threshold,
            })
            df = pd.concat([df, res])
df
````

````python
plt.rcParams["figure.figsize"] = (8, 5)
ax = sns.boxplot(data=df, x="threshold", y="rg_all", hue="protein")
ax.set_ylabel("Rg (nm)")
ax.set_xlabel(None)
ax.set_title("Radius of gyration calculated on backbone")
ax.tick_params(axis='x', labelrotation=45)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import os

dir_names = "baseline all_common 0.1nm 0.05nm 0.025nm 0.01nm no_rubber".split(" ")
label_names = ["Baseline", "No threshold", "Threshold: 0.1 nm", "Threshold: 0.05 nm", "Threshold: 0.025 nm" , "Threshold: 0.01 nm"]

df_in = pd.DataFrame()
for protein in ["5eqi", "4zw9"]:
    for threshold in dir_names:
        for i in range(3):
            path = f"gr/{protein}_{threshold}_tmin_rep{i+1}.xvg"
            if not os.path.isfile(path):
                continue

            x, y, rg_x, rg_y, rg_z = np.loadtxt(path,comments=["@", "#"],unpack=True)
            res = pd.DataFrame({
                "time" : x,
                "rg_all" : y,
                "rg_x"   : rg_x,
                "rg_y"   : rg_y,
                "rg_z"   : rg_z,
                "replicate" : f"rep {i+1}",
                "protein"   : protein,
                "threshold" : threshold,
            })
            df_in = pd.concat([df_in, res])

df_out = pd.DataFrame()
for protein in ["5eqi", "4zw9"]:
    for threshold in dir_names:
        for i in range(3):
            path = f"gr/{protein}_{threshold}_tmout_rep{i+1}.xvg"
            if not os.path.isfile(path):
                continue

            x, y, rg_x, rg_y, rg_z = np.loadtxt(path,comments=["@", "#"],unpack=True)
            res = pd.DataFrame({
                "time" : x,
                "rg_all" : y,
                "rg_x"   : rg_x,
                "rg_y"   : rg_y,
                "rg_z"   : rg_z,
                "replicate" : f"rep {i+1}",
                "protein"   : protein,
                "threshold" : threshold,
            })
            df_out = pd.concat([df_out, res])

df_out
````

````python
fig, axs = plt.subplots(1, 2)
plt.rcParams["figure.figsize"] = (10, 4)

axs[0].axhline(1.39, c="orange", alpha=0.6)
axs[0].axhline(1.55, c="darkblue", alpha=0.6)
sns.boxplot(data=df_in, x="threshold", y="rg_all", hue="protein", ax=axs[0])
axs[0].set_ylabel("Rg (nm)")
axs[0].set_xlabel("Threshold")
axs[0].set_title("Radius of gyration TM_IN")

axs[1].axhline(1.5, c="orange", alpha=0.6)
axs[1].axhline(1.41, c="darkblue", alpha=0.6)
sns.boxplot(data=df_out, x="threshold", y="rg_all", hue="protein", ax=axs[1])
axs[1].set_ylabel("Rg (nm)")
axs[1].set_xlabel("Threshold")
axs[1].set_title("Radius of gyration TM_OUT")

axs[0].tick_params(axis='x', labelrotation=45)
axs[1].tick_params(axis='x', labelrotation=45)
````
