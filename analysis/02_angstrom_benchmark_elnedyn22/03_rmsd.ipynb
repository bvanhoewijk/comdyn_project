```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

def sliding_window_average(x, y, window_size):
    x = np.array(x)
    y = np.array(y)
    weights = np.repeat(1.0, window_size) / window_size
    y_avg = np.convolve(y, weights, 'valid')
    x_avg = np.convolve(x, weights, 'valid')
    return x_avg, y_avg

def plot_rmsd2(base_path, title, protein):
    plt.rcParams["figure.figsize"] = (12,3)
       
    fig, axs = plt.subplots(1, 3)
    for i in range(0, 3):
        x, y = np.loadtxt(f"{base_path}_rep{i+1}.xvg",comments=["@", "#"],unpack=True)
        # x, y = sliding_window_average(x, y, 50)
        x /= 1e3 # ns
        y *= 10  # nm to angstrom

        y_hist,binEdges=np.histogram(y,bins=50)
        bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
        p = y_hist/float(y_hist.sum())
        p[p==0] = None

        # Joule/Kelvin
        k_b = 1.380649*(10**(-23))
        T = 303.15 # Kelvin

        # Free energy:
        avagadro = 6.022 * 10 ** 23       
        energy = -(k_b * T * np.log(p)) 

        # energy to kilojoule per mol:
        # kilo = 1000
        fe = energy * avagadro / 1000

        axs[0].plot(bincenters, fe,'-', marker=".")
        axs[0].set_xlabel("RMSD (A)") 
        axs[0].set_ylabel(r"Free Energy (kJ$\cdot$mol$^{-1}$)")
        axs[0].set_title(f"Free energy")

        axs[1].scatter(x, y, s=1)
        axs[1].set_ylabel("RMSD (A)")
        axs[1].set_xlabel("Time (ns)") 
        axs[1].set_title(f"PDB-ID: {protein}\n{title}\n\nRMSD on backbone (C-alpha)")
        axs[1].set_ylim((0, 5.5))

        y_hist, binEdges=np.histogram(y,bins=100)

        bincenters = 0.5*(binEdges[1:]+binEdges[:-1])
        axs[2].plot(bincenters, y_hist/sum(y_hist), '-', marker=".")
        axs[2].set_xlabel("RMSD (A)")
        axs[2].set_ylabel("Frequency")
        axs[2].set_title(f"RMSD histogram")
        axs[2].set_xlim((0, 5.5))

        axs[2].legend(("rep1", "rep2", "rep3"), title="Replicates")

    fig.subplots_adjust(hspace=0.6)
    fig.subplots_adjust(wspace=0.3)


protein="5eqi"
plot_rmsd2(f"rmsd/{protein}_baseline", title="Baseline", protein=protein.upper())
plot_rmsd2(f"rmsd/{protein}_all_common",    title="All common", protein=protein.upper())
plot_rmsd2(f"rmsd/{protein}_0.1nm",    title="Threshold: 0.1nm", protein=protein.upper())
plot_rmsd2(f"rmsd/{protein}_0.05nm",   title="Threshold: 0.05nm", protein=protein.upper())
plot_rmsd2(f"rmsd/{protein}_0.025nm",  title="Threshold: 0.025nm", protein=protein.upper())
plot_rmsd2(f"rmsd/{protein}_0.01nm",   title="Threshold: 0.01nm", protein=protein.upper())

protein="4zw9"
plot_rmsd2(f"rmsd/{protein}_baseline", title="Baseline", protein=protein.upper())
plot_rmsd2(f"rmsd/{protein}_all_common",    title="All common", protein=protein.upper())
plot_rmsd2(f"rmsd/{protein}_0.1nm",    title="Threshold: 0.1nm", protein=protein.upper())
plot_rmsd2(f"rmsd/{protein}_0.05nm",   title="Threshold: 0.05nm", protein=protein.upper())
plot_rmsd2(f"rmsd/{protein}_0.025nm",  title="Threshold: 0.025nm", protein=protein.upper())
plot_rmsd2(f"rmsd/{protein}_0.01nm",   title="Threshold: 0.01nm", protein=protein.upper())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

def sliding_window_average(x, y, window_size):
    x = np.array(x)
    y = np.array(y)
    weights = np.repeat(1.0, window_size) / window_size
    y_avg = np.convolve(y, weights, 'valid')
    x_avg = np.convolve(x, weights, 'valid')
    return x_avg, y_avg

paths = ["baseline", "all_common","0.1nm", "0.05nm", "0.025nm", "0.01nm", "no_rubber"]
titles = ["Baseline", "All common", "Threshold: 0.1nm", "Threshold: 0.05nm", "Threshold: 0.025","0.01nm", "No rubber bands"]

names = dict(zip(paths, titles))

plt.rcParams["figure.figsize"] = (15,7)

dataset = []
for protnr, protein in enumerate(["5eqi", "4zw9"]):
    for tnr, threshold in enumerate(paths):
        for i in range(0, 3):
            try:
                x, y = np.loadtxt(f"rmsd/{protein}_{threshold}_rep{i+1}.xvg",comments=["@", "#"],unpack=True)
                x, y = sliding_window_average(x, y, 10)
                x /= 1e3 # ns
                # y *= 10  # nm to angstrom

                for y_val in y:
                    df = {"protein" : protein,
                          "threshold": threshold,
                          "replicate": f"rep{i+1}",
                          "y": y_val}
                    dataset.append(df)
                
            except:
                print(f"BUG! \"{protein}:rep{i+1}\" Threshold:\"{threshold}\"")
                pass
    
dataset = pd.DataFrame(dataset)
```

```python
def fixname(x):
    if x == "4zw9":
        return "GLUT3"
    if x == "5eqi": 
        return "GLUT1"
    return x
    
dataset['name'] = dataset['protein'].apply(fixname)
```
